```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = file.drop("Outcome", axis =1)
y = file["Outcome"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)

```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# evaluate
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
```

```python
y_pred = model.predict(x_test)
```

```python
print(f"The F1 score is: {f1_score(y_pred = y_pred, y_true = y_test)}")
print(f"The accuracy score is: {accuracy_score(y_pred = y_pred, y_true = y_test)}")
print(f"The  Recall score is: {recall_score(y_pred = y_pred, y_true = y_test)}")
print(f"The precision score is: {precision_score(y_pred = y_pred, y_true = y_test)}")
```
